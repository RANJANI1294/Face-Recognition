```python
import cv2
import numpy as np
image=cv2.imread('index.jpg')
zeros = np.zeros((image.shape[0], image.shape[1]), np.uint8)

b,g,r = cv2.split(image)
print("Blue channel: ", b)
print("Green channel: ", g)
print("Red channel: ", r)

blue = cv2.merge([b, zeros, zeros])
green = cv2.merge([zeros, g, zeros])
red = cv2.merge([zeros, zeros, r])

custom_image = cv2.merge([b, g+100, r])
custom_image1 = cv2.merge([b+100, g, r])
custom_image2 = cv2.merge([b, g, r+100])

cv2.imwrite('BlueImage.png', blue)
cv2.imwrite('RedImage.png',red)
cv2.imwrite('GreenImage.png',green)


cv2.imshow('custom_image', custom_image)
cv2.imshow('custom_image1', custom_image1)
cv2.imshow('custom_image2', custom_image2)

cv2.waitKey(0)
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture('sample.mp4')

while cap.isOpened():
    sucess, frame = cap.read()
    print(sucess)
    if sucess:
        print("Yay!!! We got the video.")
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        cv2.imshow('Gray Frame', gray_frame)
        cv2.imshow('Frame', frame)

        k = cv2.waitKey(50)
        if k & 0xff == ord('r'):
            break
    else:
        print("Sed Lyf :( :(")
        break

cap.release()
cv2.destroyAllWindows()
```
